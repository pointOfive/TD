```python
from sklearn import metrics
```

```python
for prediction_probs in ['probs_random', 'probs_intercept', 'probs_demographic', 'probs_demographic_proc_diag', 'probs_all_available']:
    fpr = []
    tpr = []

    for thresh in np.linspace(0,1,20):
        conf_mat = metrics.confusion_matrix(y, [1 if p > thresh else 0 for p in globals()[prediction_probs]])
        fpr = fpr + [conf_mat[0][1]/float(sum(conf_mat[0]))]
        tpr = tpr + [conf_mat[1][1]/float(sum(conf_mat[1]))]
    
    plt.plot(fpr, tpr, label = prediction_probs)
    plt.ylabel("FPR (1-specificity)")
    plt.ylabel("TPR (power = sensitivity)")    
    
plt.legend(loc="best")
```

```python
# let's examine diagnoses and procedure distribution
print len(frog.procedure1.unique())
print len(frog.p_diag.unique())
print len(frog.provider.unique())

n_providers = len(frog.provider.unique())
hist_dists = np.zeros([n_providers,n_providers])
for i in range(n_providers):
    print i
    a=pd.pivot_table(frog.ix[frog.provider == frog.provider.unique()[i],['p_diag', 'procedure1', 'intercept']], values=['intercept'], index=['p_diag'], columns=['procedure1'], aggfunc=np.sum).fillna(0)
    a.reset_index()
    a.columns = a.columns.droplevel(0)
    aa_tot = sum(sum(a.values))
    for j in range(0,i):
        b=pd.pivot_table(frog.ix[frog.provider == frog.provider.unique()[j],['p_diag', 'procedure1', 'intercept']], values=['intercept'], index=['p_diag'], columns=['procedure1'], aggfunc=np.sum).fillna(0)
        b.reset_index()
        b.columns = b.columns.droplevel(0)
        b_tot = sum(sum(b.values))
        shared_procedures = [k for k in a.index.values if k in b.index.values]
        shared_diagnoses = [k for k in a.columns.values if k in b.columns.values]
        aa = a.ix[shared_procedures,shared_diagnoses].values.flatten()
        b = b.ix[shared_procedures,shared_diagnoses].values.flatten()
        hist_dist = (aa_tot-sum(aa))/aa_tot + (b_tot-sum(b))/b_tot + sum(np.abs(aa/aa_tot-b/b_tot)) 
        hist_dists[i, j] = hist_dist
```

```python
import seaborn as sns
sns.heatmap(hist_dists)
```

```python
sns.heatmap(hist_dists+hist_dists.T)
```

```python
from scipy.cluster import hierarchy as hc
provider_dists = hc.distance.squareform(hist_dists+hist_dists.T) # convert to condensed
z = hc.linkage(provider_dists, method='average')
plt.figure(figsize=[18,6])
dendrogram = hc.dendrogram(z)
plt.show()
```

```python
from scipy.cluster.hierarchy import fcluster
#max_d = 1.9
#clusters = fcluster(z, max_d, criterion='distance')
k=2
clusters = fcluster(z, k, criterion='maxclust')
print len(np.unique(clusters))
print clusters
sum(clusters==1)
```

```python
dists = hist_dists+hist_dists.T
#print clusters[np.argsort(clusters)]
sns.heatmap(dists[:, np.argsort(clusters)][np.argsort(clusters)])
# these are probably the leftmost 18 providers from the dendrogram
```
