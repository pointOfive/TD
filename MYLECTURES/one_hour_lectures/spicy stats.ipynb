```python
group_A = stats.gamma(a=2, scale=2).rvs(5000)
group_B = stats.gamma(a=1.5, scale=2.8).rvs(5000)

test_statistic = np.mean(group_A) - np.mean(group_B)
s2_A = np.var(group_A, ddof = 1)
s2_B = np.var(group_B, ddof = 1)
var_xbarA_minus_xbarB = s2_A/group_A.size + s2_B/group_B.size

xs=np.linspace(-.5,.5)
plt.plot(xs, stats.norm.pdf(xs, loc=0, scale=np.sqrt(var_xbarA_minus_xbarB)))

plt.vlines(test_statistic, ymin=0, ymax=.6)
```

```python
2*stats.norm.cdf(test_statistic, loc=0, scale=np.sqrt(var_xbarA_minus_xbarB))
```
