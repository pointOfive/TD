```python
x = np.random.normal(np.zeros([1000,1]), 10)
y = x+np.random.normal(np.zeros([1000,1]), 10)

fig, ax1 = plt.subplots(figsize=[5,5])

ax2 = ax1.twinx()
ax2.set_ylabel('Standardized data')
ax2 = ax2.twiny()
ax1.plot(x, y, '.', color='w')
ax2.plot((x-np.mean(x))/np.std(x), (y-np.mean(y))/np.std(y), '.', color='k')

ax1.set_xlabel('Unstandardized data')
ax1.set_ylabel('Unstandardized data')
ax2.set_xlabel('Standardized data')
```

```python
x = np.random.normal(np.zeros([1000,1]), 5)
y = x+np.random.normal(np.zeros([1000,1]), 5)

fig, ax1 = plt.subplots(figsize=[5,5])

plt.plot(x, (y-np.mean(y))/np.std(y), '.', color='k')
ax1.set_ylim([-20, 20])
```
