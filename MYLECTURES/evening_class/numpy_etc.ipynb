```python
import pandas as pd
```

```python
Schools_df = pd.read_csv("stuff2/Schools.csv")
Players_df = pd.read_csv("stuff2/SchoolsPlayers.csv")
```

```python
Players_df.head()
```

```python
Players_df.playerID.value_counts()
Schools_df.schoolState.value_counts()
```

```python
Schools_df.head()
```

```python
Schools_df.schoolNick.unique()
#Schools_df[Schools_df.schoolState=='MA'].schoolNick.unique()
Schools_df[Schools_df.schoolCity=='Austin'].schoolNick.unique()
#Schools_df.ix[Schools_df.schoolNick=="Lord Jeffs",2]
```

```python
Schools_df.shape
```

```python
Schools_df = Schools_df.dropna(how='any')
Players_df = Players_df.dropna(how='any')
```

```python
schoolState_dummies = pd.get_dummies(Schools_df.schoolState, prefix='my_dummy_for').astype(np.int8)
schoolState_dummies.head()
```

```python
Schools_Players_df = Schools_df.join(schoolState_dummies)
Schools_Players_df.head()
```

```python
pd.merge(Schools_df, Players_df, on='schoolID')
```

```python
Schools_Players_df = pd.merge(Schools_df, Players_df, on='schoolID')

pd.pivot_table(Schools_Players_df, values='playerID', 
               index='schoolState', columns='schoolNick',
              aggfunc = lambda x: len(x.unique()))
```
